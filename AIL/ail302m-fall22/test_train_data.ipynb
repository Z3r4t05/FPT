```python
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
test = pd.read_csv(
    r"D:\GitHub\FPT\AIL\ail302m-fall22\input\test.csv", sep=";", decimal='.')
```

```python
test_round = test.round({'volatile acidity': 1, 'density': 4})
test_round
```

```python
train = pd.read_csv(r"D:\GitHub\FPT\AIL\ail302m-fall22\input\train.csv", sep=";", decimal='.')
train
```

```python
train_raw = pd.read_csv(r"D:\GitHub\FPT\AIL\ail302m-fall22\input\train_raw.csv", sep=",", decimal='.')
train_raw_rounded = train_raw.round({'volatile acidity': 1, 'density': 4})
train_raw_rounded
```

```python
train_cols = train.columns.values
train_cols
```

```python
intersection = pd.merge(test_round, train_raw_rounded, how = 'left', on = list(test.drop('id', axis = 1).columns.values))
intersection
```

```python
intersection_drop_duplicates = intersection.drop_duplicates()
intersection_drop_duplicates
```

```python
sns.histplot(data=intersection, x='quality')
sns.scatterplot(data=intersection, x='quality')
```

```python
sns.histplot(data=train_raw, x='quality')
sns.scatterplot(data=train_raw, x='quality')
```

```python
sns.histplot(data = train, x='quality')
sns.scatterplot(data=train, x='quality')
```

```python
intersection.isnull().sum()
```

```python
nan = intersection_drop_duplicates[intersection_drop_duplicates.isnull().any(axis=1)]
nan
```

```python
nan_intersect_train = pd.merge(nan.drop('id', axis=1), train, how='left', on=list(test.drop('id', axis=1).columns.values))
nan_intersect_train
```

```python
nan_intersect_train = nan_intersect_train.drop('quality_x', axis=1).rename(columns={'quality_y': 'quality'})
nan_intersect_train
```

```python
nan_intersect_train['quality'].fillna(train['quality'].median(), inplace = True)
nan_intersect_train
```

```python
intersection_drop_duplicates.tail(1)
```

```python
nan_intersect_train.tail(1)
```

```python
final_merge = pd.merge(intersection_drop_duplicates, nan_intersect_train, how='left', on=list(test.drop('id', axis=1).columns.values))
final_merge
```

```python
final_merge = final_merge.bfill(axis = 1).drop('quality_y', axis=1).rename(columns={'quality_x': 'quality'})
final_merge
```

```python
final_merge.to_csv('input/cheating_data.csv', index=False)
```
